```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
URL = "https://www.nba.com/players"
driver.get(URL)
```

```python
# toggle 'Show Historic' button
driver.find_element(By.XPATH, '''/html/body/div[1]/div[2]/div[3]
                                 /section/div[1]/div[2]/div[1]/div[6]''').click()
```

```python
# press 'Page: All' button
driver.find_element(By.XPATH, '''/html/body/div[1]/div[2]/div[3]
                                 /section/div/div[2]/div[1]/div[7]
                                 /div/div[3]/div/label/div/select
                                 /option[1]
                                             ''').click()
```

```python
html = driver.page_source.encode('utf-8')
soup = BeautifulSoup(html, 'lxml')
```

```python
player_box = soup.find("table", "players-list")
column_labels = player_box.find("thead")

column_label = column_labels.find_all("th")
column_list = []
for label in column_label:
    column_list.append(label.text)

print(column_list)

#convert to hash table
hashed_data = {}
for column in column_list:
    hashed_data[column] = []
    
print(hashed_data)
    
player_list = player_box.find("tbody")
players = player_list.find_all("tr")
for player in players:
    attributes = player.find_all("td")
    for attribute, key in zip(attributes, hashed_data.keys()):
#         print("KEY:",key)
#         print(attribute.text)
        hashed_data[key].append(attribute.text)
        

# print(hashed_data)
```

```python
import pprint
pp = pprint.PrettyPrinter(indent=1, sort_dicts=False, compact=True, width=125)
pp.pprint(hashed_data)
```
