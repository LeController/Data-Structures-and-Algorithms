```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import time 

x = np.logspace(0,15,1000000,dtype=float) # default base=10

start_time = time.time()
y_1 = np.ones(1000000)
end_time_y_1 = time.time() - start_time

start_time = time.time()
y_n = x
end_time_y_n = time.time() - start_time 

start_time = time.time()
y_square = x*x
end_time_y_square = time.time() - start_time 

start_time = time.time()
y_squareroot = np.sqrt(x)
end_time_y_squareroot = time.time() - start_time 

start_time = time.time()
y_nlogn = x*np.log(x)
end_time_y_nlogn = time.time() - start_time 

start_time = time.time()
y_cube = x**3
end_time_y_cube = time.time() - start_time 

start_time = time.time()
y_2n = 2**x
end_time_y_2n = time.time() - start_time 
```

```python
fig = plt.figure(figsize=(10,8))
p = fig.add_subplot(1,1,1)
p.set_yscale('log')
p.set_xscale('log')
p.set_xlim(1,10**15)
p.set_ylim(1,10**19)
p.plot(x,y_1,label='1')
p.plot(x,y_n,label='n')
p.plot(x,y_square,label="n^2")
p.plot(x,y_squareroot,label="sqrt(n)")
p.plot(x,y_nlogn,label="nlogn")
p.plot(x,y_cube,label="n^3")
p.plot(x,y_2n,label="2^n")
p.legend(fancybox=True, shadow=True, title="Legend", loc="lower right", fontsize="small")
```

```python
x_labels = ['1','n','n^2','sqrt(n)','nlogn','n^3','2^n']

y_values = [end_time_y_1, end_time_y_2n, end_time_y_square, end_time_y_squareroot, end_time_y_nlogn, end_time_y_cube, end_time_y_2n]

plt.plot(x_labels, y_values)
plt.xlabel('Function')
plt.ylabel('Running time (seconds)')
plt.grid()
plt.show()
```
