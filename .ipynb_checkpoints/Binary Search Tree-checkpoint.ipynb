```python
class Node:
    def __init__(self, val):
        self.l = None
        self.r = None
        self.v = val

class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def add(self, val):
        if self.root is None:
            self.root = Node(val)
            print('added root node:', val)
        else:
            self._add(val, self.root)

    def _add(self, val, node):
        if val < node.v:
            if node.l is not None:
                self._add(val, node.l)
            else:
                node.l = Node(val)
                print('parent:',node.v,'-- added left child:', val)
        else:
            if node.r is not None:
                self._add(val, node.r)
            else:
                node.r = Node(val)
                print('parent:',node.v,'-- added right child:', val)

    def find(self, val):
        if self.root is not None:
            return self._find(val, self.root)
        else:
            return None

    def _find(self, val, node):
        if val == node.v:
            return node
        elif (val < node.v and node.l is not None):
            return self._find(val, node.l)
        elif (val > node.v and node.r is not None):
            return self._find(val, node.r)

    def deleteTree(self):
        self.root = None
        print('tree deleted')

    def printTree(self):
        print('printing tree:')
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node):
        if node is not None:
            self._printTree(node.l)
            print(str(node.v) + ' -- ')
            self._printTree(node.r)
    
    def minimum(self, val):
        node = self.find(val)
        while node.l is not None:
            node = node.l
        minimum = node.v
        return minimum
            
    def successor(self, val):
        node = self.find(val)
#         print('node.l.v:', node.l.v)
        if node.r is not None:
            return self.minimum(node.r.v)
        elif node.l is not None:
            node = node.l
            return self.successor(node.v)
        # what if node has no children, how to get parent?
#         else:
#             return self.getRoot().v

            
tree = Tree()
tree.add(3)
tree.add(4)
tree.add(0)
tree.add(8)
tree.add(2)
tree.printTree()
print(tree.find(8).v)

print('successor:',tree.successor(3))

print('MINIMUM:',tree.minimum(3))
# tree.deleteTree()
# tree.printTree()
```
